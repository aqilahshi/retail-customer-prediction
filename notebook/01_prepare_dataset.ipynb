```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("../data/analytics_sales.csv")
```

```python
df.describe()
```

```python
df["order_date"] = pd.to_datetime(df["order_date"])
df["signup_date"] = pd.to_datetime(df["signup_date"])
```

```python
print("Customers :", df["customer_id"].nunique())
print("Orders    :", df["order_id"].nunique())
print("Products  :", df["product_id"].nunique())
```

```python
customer_df = (
    df.groupby("customer_id")
      .agg(
          total_revenue=("subtotal", "sum"),
          total_orders=("order_id", "nunique"),
          average_order_value=("subtotal", "mean"),
          total_quantity=("quantity", "sum"),
          age=("age", "first"),
          gender=("gender", "first"),
          city=("city", "first"),
          membership=("membership", "first"),
          favorite_payment=("payment_method", lambda x: x.mode()[0]),
          favorite_category=("category_name", lambda x: x.mode()[0]),
          signup_date=("signup_date", "first"),
          first_purchase=("order_date", "min"),
          last_purchase=("order_date", "max")
      )
      .reset_index()
)
```

```python
customer_df.head()
```

```python
from datetime import datetime

today = df["order_date"].max()

customer_df["customer_tenure_days"] = (
    today - customer_df["signup_date"]
).dt.days

customer_df["days_since_last_purchase"] = (
    today - customer_df["last_purchase"]
).dt.days

customer_df["purchase_frequency"] = (
    customer_df["total_orders"] /
    (customer_df["customer_tenure_days"] / 30)
).round(2)

customer_df["average_items_per_order"] = (
    customer_df["total_quantity"] /
    customer_df["total_orders"]
).round(2)
```

```python
customer_df[
    [
        "customer_id",
        "total_revenue",
        "total_orders",
        "purchase_frequency",
        "customer_tenure_days",
        "days_since_last_purchase",
        "average_items_per_order",
    ]
].head()
```

```python
customer_df["total_revenue"].describe()
```

```python
threshold = customer_df["total_revenue"].quantile(0.75)

customer_df["high_value_customer"] = (
    customer_df["total_revenue"] >= threshold
).astype(int)
```

```python
customer_df["high_value_customer"].value_counts()
```

```python
customer_df.to_csv(
    "../data/customer_features.csv",
    index=False
)

print("Customer feature dataset saved successfully!")
```
